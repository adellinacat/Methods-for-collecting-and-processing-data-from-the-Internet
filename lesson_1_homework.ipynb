```python
import requests
from pprint import pprint
import json
```

```python
url = 'https://api.github.com'
user='adellinacat'
```

```python
response = requests.get(f'{url}/users/{user}/repos')
```

```python
with open('data.json', 'w') as f:
    json.dump(response.json(), f)
```

```python
for i in response.json():
    print(i['name'])
```

2. Изучить список открытых API, возьмите API вконтакте (https://vk.com/dev/first_guide).
Сделайте запрос, чтобы получить список всех сообществ на которые вы подписаны.

```python
groupsurl = 'https://api.vk.com/method/groups.get'
user = '11449151'
api_version = '5.131'
token= 'vk1.a.some simbols............'
```

```python
getgroups = requests.get(f'{groupsurl}?user_id={user}&extended=1&v={api_version}&access_token={token}')
```

```python
print('ответ: \n', getgroups.text)
```

```python
with open('vkgroups.json', 'w') as f:
    json.dump(getgroups.json(), f)
```

```python
responseitems = getgroups.json()["response"]
groupslist = responseitems["items"]
for i in groupslist:
   print(i['name'], i['screen_name'])
```
